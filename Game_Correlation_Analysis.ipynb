```python
csgo_data['Total Assists'] = csgo_data['Assists / Round'] * csgo_data['Rounds Played']
csgo_data['Total Damage'] = csgo_data['Damage / Round'] * csgo_data['Rounds Played']
```

```python
csgo_data
```

```python
# Import necessary libraries
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import ttest_ind, mannwhitneyu


# Step 1: Selecting relevant metrics (Kills, Deaths, Assists)
csgo_relevant = csgo_data[['Total Kills', 'Total Deaths', 'Total Assists']]
lol_relevant = lol_data[['kills', 'deaths', 'assists']]

# Step 2: Normalizing the data using Min-Max Scaling
scaler = MinMaxScaler()

# Normalize the CSGO data
csgo_normalized = pd.DataFrame(scaler.fit_transform(csgo_relevant), columns=csgo_relevant.columns)

# Normalize the LOL data
lol_normalized = pd.DataFrame(scaler.fit_transform(lol_relevant), columns=lol_relevant.columns)

# Correlation analysis between 'kills' from both games
csgo_normalized['kills'] = csgo_normalized['Total Kills']
lol_normalized['kills'] = lol_normalized['kills']

# Merge based on 'kills' to analyze correlations between both games
merged_data = pd.DataFrame({
    'CSGO_kills': csgo_normalized['kills'], 
    'LOL_kills': lol_normalized['kills'][:len(csgo_normalized)]  # Trim the LOL data to match the CSGO data length
})

# Step 3: Calculate the correlation between CSGO and LOL kills
correlation = merged_data.corr()
print("Correlation between CSGO and LOL kills:")
print(correlation)
```

```python
# Step 5: Running T-test or Mann-Whitney U-test for 'deaths'
csgo_deaths = csgo_normalized['Total Deaths']
lol_deaths = lol_normalized['deaths'][:len(csgo_normalized)]

# Run T-test (assuming normal distribution)
t_stat, p_val = ttest_ind(csgo_deaths, lol_deaths)
print(f"T-test results - Statistic: {t_stat}, P-value: {p_val}")
```

```python
# Run Mann-Whitney U-test (if not assuming normal distribution)
u_stat, p_val_u = mannwhitneyu(csgo_deaths, lol_deaths)
print(f"Mann-Whitney U-test results - Statistic: {u_stat}, P-value: {p_val_u}")
```

```python
# Step 6: K/D Ratio calculation (handling division by zero)
# Calculate K/D ratio for LOL, avoiding division by zero
lol_data['K/D Ratio'] = lol_data['kills'] / lol_data['deaths'].replace(0, 1)  # Replace 0 deaths with 1 to avoid division by zero

# Normalize the K/D ratio
csgo_kd_ratio = scaler.fit_transform(csgo_data[['K/D Ratio']])
lol_kd_ratio = scaler.fit_transform(lol_data[['K/D Ratio']].values)

# Step 7: Correlation for K/D ratio between both games
kd_merged_data = pd.DataFrame({
    'CSGO_KD': csgo_kd_ratio.flatten(), 
    'LOL_KD': lol_kd_ratio.flatten()[:len(csgo_kd_ratio)]
})

# Calculate correlation for K/D ratios
kd_correlation = kd_merged_data.corr()
print("Correlation between CSGO and LOL K/D Ratio:")
print(kd_correlation)
```

```python
# Step 8: Visualization for K/D ratio correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(x='CSGO_KD', y='LOL_KD', data=kd_merged_data)
plt.title("Scatter plot of CSGO vs LOL K/D Ratio (Normalized)")
plt.xlabel("CSGO K/D Ratio (Normalized)")
plt.ylabel("LOL K/D Ratio (Normalized)")
plt.show()
```
